```python
import numpy as np
import pandas as pd

# Quick sanity check
arr = np.arange(10)
df = pd.DataFrame({"value": arr, "squared": arr ** 2})
df
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(df["value"], df["squared"], marker="o")
ax.set_xlabel("x")
ax.set_ylabel("x\u00b2")
ax.set_title("Squares")
ax.grid(True)
plt.tight_layout()
plt.show()
```
